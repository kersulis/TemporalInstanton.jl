\begin{align}
\hat{B} &= B_{22} - (B_{12}^\top B_{11}^{-1} B_{12})
\end{align}

$B_{22}$ is always diagonal, so $\hat{B}$ will be diagonal iff $(B_{12}^\top B_{11}^{-1} B_{12})$ is diagonal.

Write the SVD of $B_{12}$ and the Eigendecomposition of $B_{11}^{-1}$:

\begin{align}
B_{12} &= U_{12}S_{12}V_{12}^\top & B_{11}^{-1} &= U_{11}\Lambda_{11} U_{11}^\top
\end{align}

Note that the Eigendecomposition of $B_{11}$ matches its SVD. Now write the product $(B_{12}^\top B_{11}^{-1} B_{12})$ in terms of the decompositions:

\begin{align}
B_{12}^\top B_{11}^{-1} B_{12} &= V_{12}S_{12}U_{12}^\top U_{11}\Lambda_{11}U_{11}^\top U_{12}S_{12}V_{12}^\top
\end{align}

Note that $V_{12}=I$.
