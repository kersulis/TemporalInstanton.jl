This initial check indicates that only the smallest value of the Lagrange multiplier matters. Now for a proof.

The problem is 

\begin{align*}
\min & ~~ w^\top B w + 2w^\top b \\
s.t. & ~~ w^\top w = c
\end{align*}

The Lagrangian for this problem is

\begin{align*}
\mathcal{L}(w,\lambda) &= w^\top B w + 2w^\top b + \lambda(c - w^\top w)
\end{align*}

Any optimal solution to this problem must satisfy the first-order conditions.

\begin{align*}
\frac{\partial \mathcal{L}}{\partial w} &= 0 = 2Bw + 2b - 2\lambda w \\
\frac{\partial \mathcal{L}}{\partial \lambda} &= 0 = w^\top w - c
\end{align*}

**Theorem:** If $(w_1,\lambda_1)$ and $(w_2,\lambda_2)$ satisfy the first-order conditions, and $\lambda_1 < \lambda_2$, then $w_1^\top B w_1 + 2w_1^\top b < w_2^\top B w_2 + 2w_2^\top b$

**Proof:** Because the two points satisfy $\frac{\partial \mathcal{L}}{\partial w} = 0$, we have

\begin{align*}
Bw_1 + b &= \lambda_1w_1 \\
Bw_2 + b &= \lambda_2w_2
\end{align*}

Now pre-multiplying the top equation by $w_1^\top$ and the bottom one by $w_2^\top$, we obtain:

\begin{align*}
w_1^\top Bw_1 + w_1^\top b &= \lambda_1w_1^\top w_1 \\
w_2^\top Bw_2 + w_2^\top b &= \lambda_2w_2^\top w_2
\end{align*}

Because the two points satisfy $\frac{\partial \mathcal{L}}{\partial \lambda} = 0$, we can substitute $w_1^\top w_1 = c$ and $w_2^\top w_2 = c$:

\begin{align*}
w_1^\top Bw_1 + w_1^\top b &= \lambda_1c \\
w_2^\top Bw_2 + w_2^\top b &= \lambda_2c
\end{align*}

Now $\lambda_1 < \lambda_2$ allows us to conclude:

\begin{align*}
w_1^\top Bw_1 + w_1^\top b < w_2^\top Bw_2 + w_2^\top b
\end{align*}
